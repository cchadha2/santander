# Correlation

```python
def correlation_map(df, columns, figsize=(15,10)):
    import matplotlib.pyplot as plt
    correlation = (df.loc[:, columns]).corr()

    fig, ax = plt.subplots(figsize = figsize)
    sns.heatmap(correlation, annot = True, ax = ax)
```

```python
train.iloc[:, [x for x in range(10)]].corr()
```

```python
correlation = train.corr()
```

```python
correlation.head()
```

```python
correlation['target'].sort_values(ascending=False).head(10)
```

```python
correlation['target'].sort_values(ascending=False).tail(10)
```

```python
corr_list = correlation['target'].sort_values(ascending=False).head(10).index.tolist()
```

```python
corr_list = corr_list + correlation['target'].sort_values(ascending=False).tail(10).index.tolist()
```

```python
corr_list
```

```python
correlation_map(train, corr_list, figsize=(20, 12))
```

No significant correlation with target
